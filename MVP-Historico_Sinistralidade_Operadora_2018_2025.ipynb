## 3.4 Dicionário de dados

Preencha a tabela abaixo com os principais atributos.

| Coluna | Tipo | Descrição | Será usada no modelo? | Observações |
|---|---|---|---|---|
| DATA | data | Data de referência do trimestre | sim | Ordenação temporal |
| ANO | numérica | ano da observação | opcional | Pode ser utilizada para análises exploratórias |
| TRIMESTRE_NUM | numérica | Número do trimestre (1 a 4) | sim | Auxilia na identificação da sazonalidade |
| TRIMESTRE | categórica | Identificação do trimestre no formato AAAA-Tn | Não | Utilizada apenas para visualização e interpretação |
| DESPESA_TRIMESTRAL | numérica | Despesa assistencial do trimestre | sim | Variável explicativa derivada das contas contábeis do grupo 41 |
| RECEITA_TRIMESTRAL | numérica | Receita de contraprestações do trimestre | sim | Variável explicativa derivada das contas contábeis do grupo 31 |
| SINISTRALIDADE | alvo | Relação percentual entre despesa assistencial e receita do trimestre | alvo | variável que será prevista pelo modelo |


```python
## 3.5 Construção da base analítica para o modelo

# Garantir tipos corretos
df["DATA"] = pd.to_datetime(df["DATA"])
df["CD_CONTA_CONTABIL"] = df["CD_CONTA_CONTABIL"].astype(str)

# Criar ano e trimestre
df["ANO"] = df["DATA"].dt.year
df["TRIMESTRE_NUM"] = df["DATA"].dt.quarter
df["TRIMESTRE"] = df["ANO"].astype(str) + "-T" + df["TRIMESTRE_NUM"].astype(str)

# Separar receita e despesa assistencial
receita = df[df["CD_CONTA_CONTABIL"].str.startswith("31")].copy()
despesa = df[df["CD_CONTA_CONTABIL"].str.startswith("41")].copy()

# Agrupar valores acumulados por período
receita_trim = (
    receita
    .groupby(["DATA", "ANO", "TRIMESTRE_NUM", "TRIMESTRE"], as_index=False)["VL_SALDO_FINAL"]
    .sum()
    .rename(columns={"VL_SALDO_FINAL": "RECEITA_ACUMULADA"})
)

despesa_trim = (
    despesa
    .groupby(["DATA", "ANO", "TRIMESTRE_NUM", "TRIMESTRE"], as_index=False)["VL_SALDO_FINAL"]
    .sum()
    .rename(columns={"VL_SALDO_FINAL": "DESPESA_ACUMULADA"})
)

# Juntar receita e despesa
base_modelo = receita_trim.merge(
    despesa_trim,
    on=["DATA", "ANO", "TRIMESTRE_NUM", "TRIMESTRE"],
    how="inner"
)

# Ordenar por data
base_modelo = base_modelo.sort_values("DATA").reset_index(drop=True)

# Transformar valores acumulados em valores do trimestre
base_modelo["RECEITA_TRIMESTRAL"] = (
    base_modelo
    .groupby("ANO")["RECEITA_ACUMULADA"]
    .diff()
)

base_modelo["DESPESA_TRIMESTRAL"] = (
    base_modelo
    .groupby("ANO")["DESPESA_ACUMULADA"]
    .diff()
)

# No primeiro trimestre de cada ano, o valor trimestral é igual ao acumulado
base_modelo["RECEITA_TRIMESTRAL"] = base_modelo["RECEITA_TRIMESTRAL"].fillna(
    base_modelo["RECEITA_ACUMULADA"]
)

base_modelo["DESPESA_TRIMESTRAL"] = base_modelo["DESPESA_TRIMESTRAL"].fillna(
    base_modelo["DESPESA_ACUMULADA"]
)

# Calcular sinistralidade trimestral
base_modelo["SINISTRALIDADE"] = (
    base_modelo["DESPESA_TRIMESTRAL"] / base_modelo["RECEITA_TRIMESTRAL"]
) * 100

# Selecionar colunas finais
base_modelo = base_modelo[
    [
        "DATA",
        "ANO",
        "TRIMESTRE_NUM",
        "TRIMESTRE",
        "RECEITA_TRIMESTRAL",
        "DESPESA_TRIMESTRAL",
        "SINISTRALIDADE"
    ]
]

# Visualizar resultado
base_modelo.head()

```

# 8. Treinamento e avaliação inicial

Treine o baseline e os modelos candidatos. Compare os resultados iniciais.

> **Comentário:** não use o teste para tomar muitas decisões iterativas. Se fizer muitos ajustes, use validação cruzada ou uma base de validação.


```python
# === Treinamento e avaliação inicial ===

import time
from statsmodels.tsa.arima.model import ARIMA
from sklearn.ensemble import RandomForestRegressor

results = {}
trained_models = {}
predictions = {}

def evaluate_forecasting(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

    return {
        "MAE": mae,
        "RMSE": rmse,
        "MAPE": mape
    }

# Baseline: último valor observado no treino
t0 = time.time()

last_value = y_train.iloc[-1]
y_pred_baseline = np.repeat(last_value, len(y_test))

train_time = time.time() - t0

results["Baseline_Ultimo_Valor"] = evaluate_forecasting(y_test, y_pred_baseline)
results["Baseline_Ultimo_Valor"]["train_time_s"] = round(train_time, 3)
predictions["Baseline_Ultimo_Valor"] = y_pred_baseline


# Modelo 1: ARIMA
t0 = time.time()

arima_model = ARIMA(y_train, order=(1, 0, 0))
arima_fit = arima_model.fit()

y_pred_arima = arima_fit.forecast(steps=len(y_test))

train_time = time.time() - t0

results["ARIMA"] = evaluate_forecasting(y_test, y_pred_arima)
results["ARIMA"]["train_time_s"] = round(train_time, 3)
trained_models["ARIMA"] = arima_fit
predictions["ARIMA"] = y_pred_arima


# Modelo 2: Random Forest Regressor
t0 = time.time()

rf_model = RandomForestRegressor(random_state=SEED)
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

train_time = time.time() - t0

results["RandomForestRegressor"] = evaluate_forecasting(y_test, y_pred_rf)
results["RandomForestRegressor"]["train_time_s"] = round(train_time, 3)
trained_models["RandomForestRegressor"] = rf_model
predictions["RandomForestRegressor"] = y_pred_rf


# Tabela de resultados
show_results_table(results)
```

```python
# === Otimização de hiperparâmetros ===
from sklearn.model_selection import TimeSeriesSplit, RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from scipy.stats import randint

N_ITER_SEARCH = 5

cv = TimeSeriesSplit(n_splits=3)

model_to_tune = RandomForestRegressor(random_state=SEED)

param_dist = {
    "n_estimators": randint(50, 250),
    "max_depth": randint(2, 10),
    "min_samples_split": randint(2, 8),
    "min_samples_leaf": randint(1, 5)
}

search = RandomizedSearchCV(
    estimator=model_to_tune,
    param_distributions=param_dist,
    n_iter=N_ITER_SEARCH,
    cv=cv,
    scoring="neg_mean_absolute_error",
    random_state=SEED,
    n_jobs=1,
    verbose=1
)

search.fit(X_train, y_train)

print("Melhor score na validação:", search.best_score_)
print("Melhores hiperparâmetros:", search.best_params_)
```

Apesar disso, alguns períodos apresentaram resíduos mais elevados, sugerindo que eventos específicos não foram totalmente capturados pelas variáveis disponíveis.

>Não foram identificados indícios claros de underfitting, uma vez que o modelo superou o baseline e os demais modelos avaliados. Também não foram observados sinais evidentes de overfitting durante a avaliação em dados não utilizados no treinamento. No entanto, devido ao número reduzido de observações da base, essa possibilidade não pode ser descartada completamente.

>O MAPE foi adotado como principal métrica de avaliação por permitir interpretar diretamente o erro percentual das previsões. As métricas MAE e RMSE foram utilizadas como complemento para avaliar a magnitude absoluta dos erros.

>Entre as principais limitações da solução destacam-se o tamanho reduzido da série histórica, a utilização de dados de apenas uma operadora e a ausência de variáveis externas que poderiam influenciar a sinistralidade, como inflação médica, perfil etário da carteira, mudanças regulatórias e eventos epidemiológicos. >Apesar dessas limitações, o modelo atingiu os critérios de sucesso definidos para o MVP, apresentando MAPE inferior a 15% e desempenho superior ao baseline, demonstrando potencial para apoiar análises preditivas da sinistralidade de operadoras de saúde.
